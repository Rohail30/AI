```python
#Write a program that takes input ending point for a sin and cos graph and plot both curves on the same graph

import matplotlib.pyplot as plt
import numpy as np
xstart=0
xend=int(input("Enter end point: "))
xincrement=0.1
x=np.arange(xstart,xend,xincrement)
y=np.cos(x)
z=np.sin(x)
plt.xlabel=("X AxiS")
plt.ylabel=("Y AxiS")
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

```python
#Draw TAN graph seperately
import matplotlib.pyplot as plt
import numpy as np
xstart=0
xend=int(input("Enter end point: "))
xincrement=0.1
x=np.arange(xstart,xend,xincrement)
y=np.tan(x)
plt.xlabel=("X AxiS")
plt.ylabel=("Y AxiS")
plt.plot(x,y)
plt.show()
```
