## Decision Tree

```python
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
inputs = df.drop(['Survived', 'Name', 'PassengerId', 'Ticket', 'Fare', 'Cabin'], axis = 'columns')
target = df['Survived' ]

from sklearn.preprocessing import LabelEncoder
le_embarked = LabelEncoder()
le_sex = LabelEncoder()
le_age = LabelEncoder()

inputs['Embarked_n'] = le_embarked.fit_transform(inputs['Embarked'])
inputs['Sex_n'] = le_sex.fit_transform(inputs['Sex'])
inputs['Age_n'] = le_age.fit_transform(inputs['Age'])
```

```python
inputs
```

```python
inputs_n = inputs.drop(['Sex', 'Embarked', 'Age'], axis = 'columns')
```

```python
inputs_n
```

```python
target
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()

model.fit(inputs_n, target)
```

```python
model.score(inputs_n,target)
```

```python
model.predict([[3, 1, 0, 2, 1, 28]])
```
